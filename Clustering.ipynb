```python
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Y, Ylabel = make_moons(n_samples= 300, noise=0.05, random_state= 123)
```

```python
plt.scatter(Y[:,0], Y[:,1], c=Ylabel, alpha=0.7)
plt.title('Dataset #2 : Original')
plt.show()
```

```python
X, label = make_blobs(n_samples= 300, n_features= 2, centers= 3, cluster_std= 5, random_state= 123)
```

```python
plt.scatter(X[:,0], X[:,1], c=label, alpha=0.7)
plt.title('Dataset #1 : Original')
plt.show()
```

```python
knmeans = KMeans(n_clusters= 3)
klabels = knmeans.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=klabels, cmap='viridis', alpha=0.7)
plt.title('K-Means')
plt.show()
```

```python
grupi = AgglomerativeClustering(n_clusters= 3)
glabels = grupi.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=glabels, cmap='viridis', alpha=0.7)
plt.title('Agglomerative Clustering')
plt.show()
```

```python
bdscan = DBSCAN(eps= 0.5, min_samples= 5)
blabels = bdscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=blabels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN Clustering')
plt.show()
```

```python
Ykmeans = KMeans(n_clusters= 3)
Yklabels = Ykmeans.fit_predict(Y)
```

```python
plt.scatter(Y[:, 0], Y[:, 1], c=Yklabels, cmap='viridis', alpha=0.7)
plt.title('K - Means for DataSet #2')
plt.show()
```

```python
Ygrupi = AgglomerativeClustering(n_clusters= 3)
Yglabels = Ygrupi.fit_predict(Y)
```

```python
plt.scatter(Y[:, 0], Y[:, 1], c=Yglabels, cmap='viridis', alpha=0.7)
plt.title('Aglomerative Clustering DataSet #2')
plt.show()
```

```python
Ydbscan = DBSCAN(eps= 0.5, min_samples= 5)
Yblabels = Ydbscan.fit_predict(Y)
```

```python
plt.scatter(Y[:, 0], Y[:, 1], c=Yblabels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN Clustering DataSet #2')
plt.show()
```
