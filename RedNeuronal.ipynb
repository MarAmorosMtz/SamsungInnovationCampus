```python
X_train = sequence.pad_sequences(X_train, maxlen=100)
X_test = sequence.pad_sequences(X_test, maxlen=100)
```

```python
y = np.concatenate([y_train, y_test], axis=0)

y = to_categorical(y, 46)

y_train = y[:n_train_size, :]
y_test = y[n_train_size:, :]
```

```python
y_test
```

```python
n_neurons = 100 #Numero de neuronas en cada celda de memoria de la red LSTM o RNN
                #Un mayor numero de neuronas permite a la red aprender patrones mas complejos

n_input = 100   #Dimension del espacio de embedding
```

```python
my_model = Sequential() #Crear un modelo secuencial, donde las capas se anaden una tras otra.

#Capa de Embedding:
#Convierte las palabras(representadas por indices) en vectores densos de tamano n_input
#n_words: tamano del vocabulario (cantidad de palabras unicas que el modelo conoce)
#n_input: dimension del espacio de embedding (cada palabra sera representada por un vector de esta dimension)
my_model.add(Embedding(n_words, n_input, input_length=1000))

#Capa de LSTM:
my_model.add(LSTM(units=n_neurons, return_sequences=False, input_shape=(None, n_input), activation='tanh'))

my_model.add(Dense(n_cat, activation='softmax'))
```

```python
import tensorflow as tf

my_model = Sequential() #Crear un modelo secuencial, donde las capas se anaden una tras otra.

#Capa de Embedding:
#Convierte las palabras(representadas por indices) en vectores densos de tamano n_input
#n_words: tamano del vocabulario (cantidad de palabras unicas que el modelo conoce)
#n_input: dimension del espacio de embedding (cada palabra sera representada por un vector de esta dimension)
my_model.add(Embedding(input_dim=n_words, output_dim=n_input, input_length=1000))

#Capa de LSTM:
my_model.add(LSTM(units=n_neurons, return_sequences=False, input_shape=(None, n_input), activation='tanh'))

my_model.add(Dense(n_cat, activation='softmax'))

my_model.build(input_shape=(None, 100))

my_model.summary()
```

```python
n_epochs = 20
batch_size = 20
learn_rate = 0.001
```

```python
my_optimizer = Adam(learning_rate=learn_rate)
my_model.compile(
    loss='categorical_crossentropy',
    optimizer=my_optimizer,
    metrics=['accuracy']
    )
```

```python
import numpy as np

# ... (rest of your code) ...

# Before fitting the model
X_train = np.array(X_train)

# ... (rest of your code) ...

my_summary = my_model.fit(
    X_train,
    y_train,
    epochs=n_epochs,
    batch_size=batch_size,
    validation_split=0.2,
    verbose=1
)
```

```python
plt.plot(my_summary.history['accuracy'], c="b")                                                  #Dibuja la precisión del entrenamiento en azul.
plt.plot(my_summary.history['val_accuracy'], c="g")                                            # Dibuja la precisión de la validación en verde.

plt.title('Training History')                                                                   # Título de la gráfica.
plt.ylabel('Accuracy')                                                                          # Etiqueta del eje Y: Precisión.

plt.xlabel('Epoch')                                                                             # Etiqueta del eje X: Número de épocas.

plt.legend(['Train', 'Validation'], loc='lower right')                                          # Agrega una leyenda para diferenciar las líneas.
plt.show()
```

```python
ACC = my_model.evaluate(X_test, y_test, verbose=0)[1]
# evaluate() devuelve una lista donde:
# - evaluate()[0] es la pérdida (loss).
# - evaluate()[1] es la precisión (accuracy), que guardamos en "ACC".

print("Test Accuracy: {}".format(np.round(ACC, 3)))
# Muestra la precisión en los datos de prueba con 3 decimales.
```
