```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

%matplotlib inline
```

```python
def load_boston():
  data_url = "http://lib.stat.cmu.edu/datasets/boston"

  raw_df = pd.read_csv(data_url, sep="\s+", skiprows = 22, header = None)

  data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])

  target = raw_df.values[1::2, 2]

  return {
      "data" :  data,
      "target" : target,
      "DESCR" : "Boston Dataset",
      "feature_names" : ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]
  }
```

```python
data = load_boston()
```

```python
data.keys()
```

```python
print("Claves del dataset: ", data.keys())

print("\nContenido de data (primeras 5 filas)\n", data["data"][:5])

print("\nContenido de target:\n", data["target"])

print("\nDescripcion del dataset:\n", data["DESCR"])

print("\nNombres de las caracteristicas:\n", data["feature_names"])
```

```python
X = data["data"]

header = data["feature_names"]

Y = data["target"]

Y = Y.reshape(-1, 1)
```

```python
predY = lm.predict(X)

plt.scatter(Y, predY, c="blue", s=15, alpha=0.5)
plt.xlabel("Real Price")
plt.ylabel("Predicted Price")
plt.show()
```

```python
r_squared = lm.score(X, Y)

r_squared
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=123)

print("Tamao de X_train: ", X_train.shape)
print("Tamao de X_test: ", X_test.shape)
print("Tamao de Y_train: ", Y_train.shape)
print("Tamao de Y_test: ", Y_test.shape)
```

```python
lm = LinearRegression(fit_intercept=True)

lm.fit(X_train, Y_train)

Y_pred_train = lm.predict(X_train)

Y_pred_test = lm.predict(X_test)
```

```python
print("In-sample MSE is : " + str(metrics.mean_squared_error(Y_train, Y_pred_train)))

print("Out-of-sample MSE is : " + str(metrics.mean_squared_error(Y_test, Y_pred_test)))

print("-"*50)

print("In-sample RMS is : " + str(np.sqrt(metrics.mean_squared_error(Y_train, Y_pred_train))))

print("Out-of-sample RMS is : " + str(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred_test))))
```

```python
residual = Y_train - Y_pred_train

plt.scatter(Y_train, residual, c="red", s=15, alpha=0.5)
plt.xlabel("Real Price")
plt.ylabel("Residual")
plt.show()
```

```python
sns.histplot(residual, bins=50, color="green").set_title("Residual Histogram")

plt.show()
```
